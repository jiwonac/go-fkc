# import pymongo
import matplotlib.pyplot as plt

```python
def plot_neighbors_hist(db_name, collection_name):
    # Connect to MongoDB
    client = pymongo.MongoClient()
    db = client[db_name]
    collection = db[collection_name]

    # Initialize list to hold counts
    true_counts = []

    # Iterate over entries in the collection
    for entry in collection.find():
        # Get the "neighbors" field
        neighbors = entry.get("neighbors", [])
        # Count the number of True values in the array
        true_count = sum(1 for x in neighbors if x)
        # Add the count to the list
        true_counts.append(true_count)

    # Plot histogram of the counts
    plt.hist(true_counts, bins=range(max(true_counts)+2))
    plt.xlabel("Number of True values")
    plt.ylabel("Count")
    plt.title("Histogram of True value counts")
    plt.show()
```

```python
plot_neighbors_hist("dummydb", "n100000d3m5r10")
```

```python
def plot_true_count(db_name, collection_name):
    # Connect to MongoDB
    client = pymongo.MongoClient()
    db = client[db_name]
    collection = db[collection_name]

    # Initialize list to hold counts
    true_counts = []

    # Iterate over entries in the collection
    for entry in collection.find():
        # Get the "neighbors" field
        neighbors = entry.get("neighbors", [])
        # Count the number of True values in the array
        true_count = sum(1 for x in neighbors if x)
        # Add the count to the list
        true_counts.append(true_count)

    # Sort the counts from highest to lowest
    true_counts.sort(reverse=True)

    # Plot a line graph of the sorted counts
    plt.plot(true_counts)
    plt.xlabel("Entries")
    plt.ylabel("Number of True values")
    plt.title("Line graph of True value counts")
    plt.show()
```

```python
plot_true_count("dummydb", "n100000d3m5r10")
```
